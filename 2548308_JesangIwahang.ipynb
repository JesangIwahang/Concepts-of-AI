**Problem 2: Advanced HDI Exploration**

```python
#Task1. Create South Asia Subset
south_asia = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
```

```python
df_south_asia = df[df['country'].isin(south_asia)]
df_south_asia.head()
```

```python
df_south_asia.to_csv("HDI_SouthAsia.csv", index=False)
```

```python
#Task2. Composite Development Score
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_south_asia[["Life_Expectancy_Index", "GNI_Index"]] = scaler.fit_transform(
    df_south_asia[["life_expectancy", "gross_inc_percap"]])


```

```python
df_south_asia["Composite_Score"] = (
    0.30 * df_south_asia["Life_Expectancy_Index"] +
    0.30 * df_south_asia["GNI_Index"])

```

```python
country_composite = (
    df_south_asia
    .groupby("country", as_index=False)
    .agg({
        "Composite_Score": "mean",
        "hdi": "mean"
    })
)

# Rankings
country_composite["Composite_Rank"] = country_composite["Composite_Score"].rank(ascending=False)
country_composite["HDI_Rank"] = country_composite["hdi"].rank(ascending=False)

country_composite
```

```python
top5 = country_composite.sort_values("Composite_Score", ascending=False).head(5)

plt.figure(figsize=(9, 5))
bars = plt.barh(top5["country"], top5["Composite_Score"], color="steelblue",edgecolor="black")

plt.xlabel("Composite Development Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Development Score (2020â€“2022 Average)")
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()
```

```python
#Overview Analysis:
#The Composite Score ranks Maldives highest, followed by Sri Lanka and Bhutan, similar to their HDI rankings.
#Differences arise because the Composite Score highlights factors like recent progress differently.
#This shows that income, health, and education contribute unevenly across South Asian countries, affecting their rankings.
#For example, Bhutanâ€™s relatively strong Composite Score might reflect specific social or environmental policies not fully captured by HDI.
```

```python
#Task3. Outlier Detection
```

```python
# HDI IQR
Q1_hdi = df_south_asia["hdi"].quantile(0.25)
Q3_hdi = df_south_asia["hdi"].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

# GNI IQR
Q1_gni =df_south_asia["gross_inc_percap"].quantile(0.25)
Q3_gni = df_south_asia["gross_inc_percap"].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

df_south_asia["Outlier"] = (
    (df_south_asia["hdi"] < Q1_hdi - 1.5 * IQR_hdi) |
    (df_south_asia["hdi"] > Q3_hdi + 1.5 * IQR_hdi) |
    (df_south_asia["gross_inc_percap"] < Q1_gni - 1.5 * IQR_gni) |
    (df_south_asia["gross_inc_percap"] > Q3_gni + 1.5 * IQR_gni)
)
```

```python
plt.figure(figsize=(8, 5))

plt.scatter(
    df_south_asia.loc[~df_south_asia["Outlier"], "gross_inc_percap"],
    df_south_asia.loc[~df_south_asia["Outlier"], "hdi"],
    alpha=0.6,
    label="Normal",
    color="skyblue"
)

plt.scatter(
    df_south_asia.loc[df_south_asia["Outlier"], "gross_inc_percap"],
    df_south_asia.loc[df_south_asia["Outlier"], "hdi"],
    label="Outliers",
    color="tomato"
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (South Asia)")
plt.legend()
plt.show()
```

```python
# The identified countries stand out as outliers because they have significantly higher GNI per capita and HDI compared to the rest of South Asia.
# This suggests stronger economies, better education, healthcare, and living standards, making their development levels noticeably different from neighboring countries.
```

```python
def scatter_with_trend(df, x_col, title):
    x = df[x_col]
    y = df["hdi"]
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, alpha=0.6, color="skyblue", edgecolor="black")
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x + b, color="red", linestyle="--", linewidth=2)
    plt.xlabel(x_col)
    plt.ylabel("HDI")
    plt.title(title)
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()

# Scatter plots for South Asia
scatter_with_trend(south_asia_2020_2022, "life_expectancy", "HDI vs Life Expectancy (South Asia)")

# Scatter plots for Middle East
scatter_with_trend(middle_east_2020_2022, "life_expectancy", "HDI vs Life Expectancy (Middle East)")

```

```python
#Task7. Outlier Detection
```

```python
def detect_outliers(df):
    Q1_hdi = df["hdi"].quantile(0.25)
    Q3_hdi = df["hdi"].quantile(0.75)
    IQR_hdi = Q3_hdi - Q1_hdi

    Q1_gni = df["gross_inc_percap"].quantile(0.25)
    Q3_gni = df["gross_inc_percap"].quantile(0.75)
    IQR_gni = Q3_gni - Q1_gni

    return (
        (df["hdi"] < Q1_hdi - 1.5 * IQR_hdi) |
        (df["hdi"] > Q3_hdi + 1.5 * IQR_hdi) |
        (df["gross_inc_percap"] < Q1_gni - 1.5 * IQR_gni) |
        (df["gross_inc_percap"] > Q3_gni + 1.5 * IQR_gni)
    )

south_asia_2020_2022["Outlier"] = detect_outliers(south_asia_2020_2022)
middle_east_2020_2022["Outlier"] = detect_outliers(middle_east_2020_2022)
```

```python
def plot_outliers(df, title):
    plt.figure(figsize=(8, 5))

    # Normal points
    plt.scatter(
        df.loc[~df["Outlier"], "gross_inc_percap"],
        df.loc[~df["Outlier"], "hdi"],
        alpha=0.6,
        color="skyblue",
        edgecolor="black",
        label="Normal"
    )

    # Outliers
    plt.scatter(
        df.loc[df["Outlier"], "gross_inc_percap"],
        df.loc[df["Outlier"], "hdi"],
        alpha=0.8,
        color="tomato",
        edgecolor="black",
        label="Outlier"
    )

    # Trend line (using all points)
    x = df["gross_inc_percap"]
    y = df["hdi"]
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m * x + b, linestyle="--", linewidth=2, label="Trend Line")

    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.title(title)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()


# Plot for South Asia
plot_outliers(south_asia_2020_2022, "Outliers in South Asia (HDI & GNI per Capita)")

# Plot for Middle East
plot_outliers(middle_east_2020_2022, "Outliers in Middle East (HDI & GNI per Capita)")

```

```python
# The plots show that most countries follow regional GNIâ€“HDI trends, but outliers stand out.
# In South Asia, some countries have high income with moderate HDI, while others achieve higher HDI despite lower income.
# In the Middle East, outliers highlight exceptionally wealthy countries, showing that income alone doesnâ€™t fully determine human development.
```
